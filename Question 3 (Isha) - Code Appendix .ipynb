```python
df_clean = df.loc[:,['final_time', 'year', 'gender']].dropna()
```

```python
df_clean = df_clean[df_clean['final_time'] != "PND"]
df_clean
```

```python
df_clean['final_time'] = pd.to_datetime(df_clean['final_time'])
```

```python
df_clean['normal'] = df_clean['final_time'].dt.normalize()
```

```python
df_clean
```

```python
df_clean['time'] = df_clean['final_time'] - df_clean['final_time'].dt.normalize()
```

```python
df_clean
```

```python
df_female = df_clean[df_clean['gender'] == "F"]
```

```python
df_male = df_clean[df_clean['gender'] == "M"]
```

```python
df_clean.groupby(['gender', 'year'])['final_time'].mean()
```

```python
df_male_ave = df_male.groupby('year')['final_time'].mean()
```

```python
df_female_ave = df_female.groupby('year')['final_time'].mean()
```

```python
male = df_male_ave.to_frame()
male
```

```python
female = df_female_ave.to_frame()
female
```

```python
male['seconds'] = pd.to_timedelta(male['final_time'].dt.strftime('%H:%M:%S')).dt.total_seconds().astype(int)
```

```python
female['seconds'] = pd.to_timedelta(female['final_time'].dt.strftime('%H:%M:%S')).dt.total_seconds().astype(int)
```

```python
female
```

```python
male
```

```python
male['seconds'].values
```

```python
year_vals = male.index.values 
```

```python
plt.plot(year_vals, male['seconds'].values)
```

```python
male['year_vals'] = year_vals
```

```python
import matplotlib.ticker as tkr
import datetime

def func(x, pos):
    return str(datetime.timedelta(seconds=x))
fmt = tkr.FuncFormatter(func)


ax = male.plot(x='year_vals', y='seconds',
               title = "Average Male Marathon Time from 2001 to 2023 for Chicago Marathon")
#ax.xaxis.set_major_formatter(fmt)
ax.yaxis.set_major_formatter(fmt)
ax.set_ylabel("Average Time")
ax.set_xlabel("Year", fontsize = 20)
ax.grid(visible=True)
ax.legend("")
```
